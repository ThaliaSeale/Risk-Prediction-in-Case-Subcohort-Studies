## Fitting Model

### Unweighted Cox PH with PCA

```python
# creating a single case subcohort dataframe
case_subcohort_df = X_tr.join(y_tr).drop(columns = "subcohort").drop_duplicates()
case_subcohort_df
```

```python
# creating the model and fitting the data
cph = CoxPHFitter()
cph.fit(case_subcohort_df, duration_col = "duration", event_col = "event")
```

`cancer_site_1.0` has low variance and is causing convergence issues. Further investigation to determine if this is adequate for prediction.

```python
sum(case_subcohort_df["cancer_site_1.0"])
```

There are only two cases of in this category, so this looks to be more of a dimensionality issue.

Let us try dimension reduction to remedy the issue.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 0.9)
X_tr_pca = pd.DataFrame(pca.fit_transform(X_tr))
X_tr_pca.index = X_tr.index
```

```python
# creating a single case subcohort dataframe
case_subcohort_df_pca = X_tr_pca.join(y_tr).drop(columns = "subcohort").drop_duplicates()
case_subcohort_df_pca
```

```python
pd.DataFrame(pca.fit_transform(X_tr))
```

```python
# creating the model and fitting the data
cph = CoxPHFitter()
cph.fit(case_subcohort_df_pca, duration_col = "duration", event_col = "event")
```

```python
cph.print_summary()
```

So the model runs if we perform dimension reduction. We need to select a suitable number of PCs. Let us use grid search.

```python
from lifelines.utils import k_fold_cross_validation

cox_pca_scores = []

for i in range(1,20):
    pca = PCA(n_components = i/20)
    X_tr_pca = pd.DataFrame(pca.fit_transform(X_tr))
    X_tr_pca.index = X_tr.index

    # creating a single case subcohort dataframe
    case_subcohort_df_pca = X_tr_pca.join(y_tr).drop(columns = "subcohort").drop_duplicates()
    
    cox_pca_scores.append(
        np.mean(k_fold_cross_validation(cph, case_subcohort_df_pca, duration_col = "duration", event_col="event", k=5, scoring_method='concordance_index'))
    )
    
print(cox_pca_scores)
print("Best variation percentage:",np.where(cox_pca_scores == np.max(cox_pca_scores))[0]/20)
```

```python
cox_pca_scores = []

for i in range(1,20):
    pca = PCA(n_components = i/20)
    X_tr_pca = pd.DataFrame(pca.fit_transform(X_tr))
    X_tr_pca.index = X_tr.index

    # creating a single case subcohort dataframe
    case_subcohort_df_pca = X_tr_pca.join(y_tr).drop(columns = "subcohort").drop_duplicates()
    
    cox_pca_scores.append(
        np.mean(k_fold_cross_validation(cph, case_subcohort_df_pca, duration_col = "duration", event_col="event", k=5, scoring_method='concordance_index'))
    )
    
print(cox_pca_scores)
print("Best variation percentage:",np.where(cox_pca_scores == np.max(cox_pca_scores))[0]/20)
```

So it seems like 0.65 of the variance has the best concordance scores. The final model:

```python
pca = PCA(n_components = 0.65)
X_tr_pca = pd.DataFrame(pca.fit_transform(X_tr))
X_tr_pca.index = X_tr.index

# creating a single case subcohort dataframe
case_subcohort_df_pca = X_tr_pca.join(y_tr).drop(columns = "subcohort").drop_duplicates()

# creating the model and fitting the data
cph = CoxPHFitter()
cph.fit(case_subcohort_df_pca, duration_col = "duration", event_col = "event")
```

```python
# Saving Model
pickle.dump(cph, open('unweighted_cox_pca.pkl', 'wb'))
```
