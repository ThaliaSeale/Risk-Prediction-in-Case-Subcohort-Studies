# Building Olink Dataset

## Knitting
Knitting the the different csv's together into a single dataset.

```python
import pandas as pd
import numpy as np
import random
random.seed(8)
```

```python
pd.set_option('display.max_columns', None)
```

Loading questionnaire data;

```python
baseline_questionnaires = pd.read_csv("D:/DAR-2022-00229-V1/data_baseline_questionnaires.csv")
```

```python
baseline_questionnaires = (baseline_questionnaires.drop(columns = ["study_date_hour","study_date_time","study_date_year","study_date_month","study_date_day",
                                       "dob_anon","dob_y","dob_m","dob_d_anon", "censoring_date","censoring_reason","has_metabolomic_data"])
     .set_index("csid"))
```

Loading olink data:

```python
baseline_olink = pd.read_csv("D:/DAR-2022-00229-V1/data_baseline_olink.csv").set_index("csid").dropna(thresh = 2)
```

```python
metabolomics = baseline_olink.join(baseline_questionnaires).drop(columns = 'plate_id')
metabolomics
```

Loading endpoints data:

```python
endpoints = pd.read_csv("D:/DAR-2022-00229-V1/endpoints.csv").set_index("csid")
# removing . from columns
endpoints.columns = endpoints.columns.map(lambda x: x.replace('.', ''))
```

Joining endpoints to data set:

```python
metabolomics = metabolomics.join(endpoints[["ep_CKB0020_combined_ep","ep_CKB0020_combined_datedeveloped"]])
metabolomics
```

Removing reading flags:

```python
metabolomics = metabolomics.filter(regex = r"^(?:(?!flag).)*$\r?\n?")
```

## Encoding
Encoding the categorical variables.

Finding which variables have more than two levels:

```python
# Loading data dataset containing value definitions
value_definitions =  pd.read_csv("D:/DAR-2022-00229-V1/value_definitions.csv")
# Finding variable types that have more than two levels
cat_vars = (value_definitions["value_type"].value_counts()
            .iloc[np.where(value_definitions["value_type"].value_counts() > 2)]
            .index)
cat_vars = np.array(cat_vars)
cat_vars
```

```python
# Loading information on what entry type each column is
column_details =  pd.read_csv("D:/DAR-2022-00229-V1/column_details.csv")
```

```python
# Finding list of columns that have  more than two levels
cat_cols = column_details.iloc[np.where(column_details["value_type"].isin(cat_vars))].column_name
cat_cols
```

```python
# Finding the columns in the metabolomics data set which are multilevel
cat_cols = metabolomics.columns[np.where(metabolomics.columns.isin(cat_cols))]
cat_cols
```

Making dummy variables:

```python
metabolomics = pd.get_dummies(metabolomics, columns = cat_cols)
metabolomics
```

## Splitting Data
We now remove a test data set as any further transformations will cause data leak.

```python
from sklearn.model_selection import train_test_split
```

```python
n_test_cases = 100
```

```python
cases_train, cases_test = train_test_split(metabolomics.query("ep_CKB0020_combined_ep == 1"),test_size = n_test_cases)
cases_train
```

```python
cases_test
```

```python
subcohort_train, subcohort_test = train_test_split(metabolomics,train_size = len(metabolomics.query("ep_CKB0020_combined_ep == 1")) - n_test_cases)
subcohort_train
```

```python
subcohort_test
```

Saving test and train datasets:

```python
cases_train.to_csv("D:/compiled_data/cases_train.csv")
cases_test.to_csv("D:/compiled_data/cases_test.csv")
subcohort_train.to_csv("D:/compiled_data/subcohort_train.csv")
subcohort_test.to_csv("D:/compiled_data/subcohort_test.csv")
```
