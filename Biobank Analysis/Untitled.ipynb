# Building Dataset

```python
import pandas as pd
import numpy as np
import random
random.seed(8)
```

```python
baseline_questionnaires = pd.read_csv("./DAR-2022-00229-V1/data_baseline_questionnaires.csv")
```

```python
baseline_questionnaires = (baseline_questionnaires.drop(columns = ["study_date_hour","study_date","study_date_time","study_date_year","study_date_month","study_date_day",
                                       "dob_anon","dob_y","dob_m","dob_d_anon", "censoring_date","censoring_reason","has_metabolomic_data"])
     .set_index("csid"))
```

```python
baseline_metabolomics = pd.read_csv("./DAR-2022-00229-V1/data_baseline_metabolomics.csv").set_index("csid").dropna(thresh = 2)
```

```python
metabolics = baseline_metabolomics.join(baseline_questionnaires)
metabolics
```

```python
endpoints = pd.read_csv("./DAR-2022-00229-V1/endpoints.csv").set_index("csid")
# removing . from columns
endpoints.columns = endpoints.columns.map(lambda x: x.replace('.', ''))
```

```python
metabolics = metabolics.join(endpoints[["ep_CKB0020_combined_ep","ep_CKB0020_combined_datedeveloped"]])
metabolics
```
