```python
import pandas as pd
import numpy as np
```

```python
model_scores = pd.read_csv("simple_linear_simulation.csv",index_col = 0)
```

```python
model_scores
```

```python
model_scores.assign(
    model = np.where(model_scores["model"] == "Cox PH Lasso, Barlow weights",
                    "Cox PH Ridge, Barlow weights",
                    np.where(model_scores["model"] == "Cox PH Ridge, Barlow weights",
                        "Cox PH Lasso, Barlow weights",
                        model_scores["model"]))
).to_csv("simple_linear_simulation.csv")
```
